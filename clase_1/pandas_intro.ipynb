### Apply

Mediante apply podemos aplicar una función definida aparte a nuestro set de datos

```python
def rate_height(height):
    if height >= 200:
        return "Tall"
    else:
        return "Not tall"
```

```python
altos = df['Height'].apply(rate_height)
altos
```

```python
df['Tallness'] = df['Height'].apply(rate_height)
```
